```python
read_file = pd.read_excel (r'C:\Users\karan\Final Table.xlsx')
read_file.to_csv (r'C:\Users\karan\PCA.csv', index = None, header=True)
df = pd.read_csv(r'C:\Users\karan\PCA.csv')

df
```

```python
from sklearn.preprocessing import StandardScaler
features = ['Shot Location x', 'Shot Location y', 'Distance', 'Shot Outcome x', 'Shot Outcome y', 'Shot Outcome z']
# Separating out the features
a = df.loc[:, features].values
# Separating out the target
b = df.loc[:,['Target']].values
# Standardizing the features
a = StandardScaler().fit_transform(a)

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(a)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, df[['Target']]], axis = 1)
finalDf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(principalComponents, np.ravel(b))
```

```python
clf.predict([[0.78, -1.3]])
```
